```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
city_data_to_load = "./Resources/city_data.csv"
ride_data_to_load = "./Resources/ride_data.csv"
```

```python
# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

```python
city_data_df
```

```python
ride_data_df
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on="city")
pyber_data_df
```

# Challenge-1.  (summary tables) 

```python
#  1. Get the total rides for each city type
total_rides_by_type = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides_by_type
```

```python
# 2. Get the total drivers for each city type
total_drivers_by_type = city_data_df.groupby(["type"]).sum()["driver_count"]
total_drivers_by_type
```

```python
#  3. Get the total amount of fares for each city type
total_fare_by_type = pyber_data_df.groupby(["type"]).sum()["fare"]
total_fare_by_type
```

```python
#  4. Get the average fare per ride for each city type
avg_fare_per_ride = total_fare_by_type / total_rides_by_type
avg_fare_per_ride
```

```python
# 5. Get the average fare per driver for each city type
avg_fare_per_driver = total_fare_by_type / total_drivers_by_type
avg_fare_per_driver
```

```python
#  6. Create a PyBer summary DataFrame
summary = {
    "Total Rides": total_rides_by_type,
    "Total Drivers": total_drivers_by_type,
    "Total Fares": total_fare_by_type,
    "Average Fare per Ride": avg_fare_per_ride,
    "Average Fare per Driver": avg_fare_per_driver
}
pyber_ride_summary_df = pd.DataFrame(summary)
pyber_ride_summary_df 
```

```python
#  7. Cleaning up the DataFrame - Delete the index name
#del pyber_ride_summary_df_index.name 
pyber_ride_summary_df.index.name = ""
pyber_ride_summary_df
```

```python
#  8. Format the columns 
pyber_ride_summary_df["Total Rides"] = pyber_ride_summary_df["Total Rides"].map("{:,}".format)
pyber_ride_summary_df["Total Drivers"] = pyber_ride_summary_df["Total Drivers"].map("{:,}".format)
pyber_ride_summary_df["Total Fares"] = pyber_ride_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_ride_summary_df["Average Fare per Ride"] = pyber_ride_summary_df["Average Fare per Ride"].map("{:,}".format)
pyber_ride_summary_df["Average Fare per Driver"] = pyber_ride_summary_df["Average Fare per Driver"].map("{:,}".format)
```

```python
pyber_ride_summary_df
```

# Challenge 2 - create a multiple line plot that shows the total weekly of the fares for each type of city 

```python
# 1. Read the merged DataFrame
# city	date	fare	ride_id	driver_count	type
pyber_data_df = pyber_data_df.rename(columns={
    "city": "City",
    "date": "Date",
    "fare": "Fare",
    "ride_id": "Ride id",
    "driver_count": "No. Drivers",
    "type": "City Type"      
})
pyber_data_df
```

```python
# New DataFrame showing the sum of the fares Using groupby()
pyber_data_df.set_index(pyber_data_df["Date"],inplace=True)
pyber_data_df
```

```python
# New DataFrame for fares only for Date City Type and Fare Columns 
column_names = ["Date", "City Type", "Fare"] 
pyber_cities_fare = pyber_data_df[column_names].copy()
pyber_cities_fare
```

```python
# Pivot table with the 'date' as the index, the columns ='City Type', and values='fare' 
pyber_cities_fare.drop(["Date"], axis=1, inplace=True)
```

```python
pyber_cities_fare.index = pd.to_datetime(pyber_data_df.index)
pyber_cities_fare.head()
```

```python
pyber_cities_fare.info()
```

```python
# Sum of fares by city and date using groupby ()
sum_fare_by_type = pyber_cities_fare.groupby(["City Type", "Date"]).sum()["Fare"]
sum_fare_by_type
```

```python
# Reset index for pivot function in pandas 
sum_fare_by_type = sum_fare_by_type.reset_index()
sum_fare_by_type
```

```python
# Pivot table with date as index and columns as city type with fare 
sum_fare_by_type_pivot = sum_fare_by_type.pivot(index="Date", columns="City Type")["Fare"]
sum_fare_by_type_pivot.head(10)
```

```python
sum_fare_by_type_pivot.info()
```

# Line chart showing fares from January 1, 2018 - April 29, 2018

```python
fare_Jan_April = sum_fare_by_type_pivot.loc['2019-01-01': '2019-04-29']
fare_Jan_April.head(10)
```

```python
# New DataFrame for function of resample by week; sum of fares for each week 
weekly_fares_df = fare_Jan_April.resample("W").sum()
weekly_fares_df
```

```python
# Graph of resampled DataFrame plotting to add titles and labels 
from matplotlib import style 
import matplotlib.pyplot as plt 
style.use("fivethirtyeight")
ax =weekly_fares_df.plot(figsize=(20,6))
#Add title 
ax.set_title("Total Fare by City Type")
# Add an x- and y-axis labels
ax.set_xlabel("Month")
ax.set_ylabel("Fare($USD)")
plt.savefig("PyBer_fare_summary.png")
plt.show 
```
